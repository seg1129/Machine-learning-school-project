```python
from data_prep2 import Data_prep2
from ml_models import project_models
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
# prep data

data_prep_instance = Data_prep2()
arrest_data_with_race = data_prep_instance.getPreprocessedArrestData()
arrest_data_no_race = data_prep_instance.getPreprocessedArrestDataWithoutRace()

w_race_x = arrest_data_with_race[:,:-1]# our features
w_race_y = arrest_data_with_race[:,-1]# our y_hat

xr_train, xr_test, yr_train, yr_test = train_test_split(w_race_x, w_race_y, test_size = 0.3, random_state = 42)

wo_race_x = arrest_data_no_race[:,:-1] # our features
wo_race_y = arrest_data_no_race[:,-1] # our y_hat

x_no_r_train, x_no_r_test, y_no_r_train, y_no_r_test = train_test_split(wo_race_x, wo_race_y, test_size = 0.3, random_state = 42)


```

```python
# train all models using data with the race feature
# project_models_instance = project_models()

trained_bayes_model = project_models.train_bayes_model(arrest_data_with_race)
trained_logistic_regres_model = project_models.train_logistic_regression(arrest_data_with_race)
trained_decision_tree_model = project_models.train_decision_tree(arrest_data_with_race)

```

```python
# ROC Graph with data using the race feature

graph = Data_prep2.Plot_ROC_ForAllModels(trained_bayes_model, trained_decision_tree_model, trained_logistic_regres_model,xr_test,yr_test)
graph.show()
```

```python
# Bayes additional analysis

bayes_predic = trained_bayes_model.predict(xr_test)
print(bayes_predic)

p_score = precision_score(yr_test, bayes_predic, average='binary', zero_division=0)
print("Bayes precision score: ", p_score)

b_recall_score_b = recall_score(yr_test, bayes_predic, average='binary', zero_division=0)
print("bayes recall score: ", b_recall_score_b)

f_score = f1_score(yr_test, bayes_predic, average='binary', zero_division=0)
print("bayes f1 score: ", f_score)
# f_score_list.append(f1_score(yhat_df, values, average='binary', zero_division=0))

# acc_score_list.append(accuracy_score(yhat_df, values))
```
